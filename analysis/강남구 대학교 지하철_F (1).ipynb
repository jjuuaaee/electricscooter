```python
import folium
from folium import Marker, Icon, CircleMarker, Circle

import pandas as pd
import csv, time
import json
```

```python
df = pd.read_csv('강남구 대학교 지하철 (상위10).csv', encoding = 'cp949')
print('='*80)

df_sample = df[['NAME','LAT','LONG','TYPE']]
display(df_sample.head(10))
df_drop_sample = df_sample.dropna(axis=0)
df_final = df_drop_sample
display(df_final.isnull().sum())
```

```python
df_sample = df[['NAME','LAT','LONG','TYPE']]
display(df_sample.head(10))
df_drop_sample = df_sample.dropna(axis=0)
df_final = df_drop_sample
display(df_final.isnull().sum())
```

```python
def seoul_map(default_location=[37.49183069512663, 127.00765659503041], default_zoom_start=12):
    base_map = folium.Map(location=default_location, 
                          control_scale=True, 
                          zoom_start=default_zoom_start)
    
    for row in df_final.itertuples():
        NAME,LAT,LONG,TYPE = row[1:]
        

        if TYPE == 'S':
            icon = Icon(color='green',icon='info-sign')
        elif TYPE == 'U':
            icon = Icon(color='pink',icon='info-sign')
            Circle(location=[LAT , LONG],
            fill=False,
            color='black',
            weight=2,
            radius=1500).add_to(base_map)


        else:
                break
            

        Marker(location=[LAT , LONG], 
               popup=f'NAME : {NAME}',
               icon = icon).add_to(base_map)
                    
    

    base_map.save('C:/Users/lokiz/강남구 대학교 지하철 map.html')
    return base_map


print('='*100)
print('강남구 대학교 지하철 map')
print('='*100)
seoul_map()
```
