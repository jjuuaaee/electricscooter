```python
import folium
from folium import Marker, Icon

import pandas as pd
import csv, time
import json
```

```python
df = pd.read_csv('서울시_전동킥보드 주차장.csv', encoding = 'cp949')
print('='*80)
print('서울시 전동킥보드 주차장')
display(df.head(10))
```

```python
df_sample = df[['NAME','LAT','LONG','TYPE']]
display(df_sample.head(10))
df_drop_sample = df_sample.dropna(axis=0)
df_final = df_drop_sample
display(df_final.isnull().sum())
```

```python
def seocho_map(default_location=[37.49183069512663, 127.00765659503041], default_zoom_start=13):
    base_map = folium.Map(location=default_location, 
                          control_scale=True, 
                          zoom_start=default_zoom_start)
    
    for row in df_final.itertuples():
        NAME,LAT,LONG,TYPE = row[1:]
        

        if TYPE == 'G':
            icon = Icon(color='red',icon='info-sign')
        elif TYPE == 'F':
            icon = Icon(color='blue',icon='info-sign')
        elif TYPE == 'S-P':
            icon = Icon(color='pink', icon='info-sign')
        elif TYPE == 'P':
            icon = Icon(color='green', icon='info-sign')
        elif TYPE == 'E':
            icon = Icon(color='beige', icon='info-sign')
        else:
                break
            

        Marker(location=[LAT , LONG], 
               popup=f'NAME : {NAME}',
               icon = icon).add_to(base_map)
        
    

    base_map.save('C:/Users/lokiz/서울시_전동킥보드 주차장 map.html')
    return base_map


print('='*100)
print('서울시_전동킥보드 주차장 map')
print('='*100)
seocho_map()
```
